```python
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np 
```

```python
#generating descriptive statistics for the dataframe 
df.describe()
```

```python
bins = [0, 100000, 300000,float('inf')]
labels = ['Low', 'Med', 'High']
df['Income Band'] = pd.cut(df['Estimated Income'],bins = bins, labels = labels , right = False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
#examine the distribution of unique categories in categorical columns 
categorical_cols = df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns

for col  in categorical_cols:
    print(f"Value Counts for'{col}':")
    display(df[col].value_counts()) 
```

```python
#UNIVARIATE ANALYSIS
for i, predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification",
"Properties Owned","Risk Weighting","Income Band"]].columns):
 plt.figure(i)
 sns.countplot(data=df,x = predictor)
```

```python
#BIVARIATE ANALYSIS
for i, predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification",
"Properties Owned","Risk Weighting","Income Band"]].columns):
 plt.figure(i)
 sns.countplot(data=df,x = predictor,hue="GenderId")
```

```python
#histogram plot of value count for different occupation 
for col in categorical_cols:
    if col == "occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of occupation Count')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
    
```

```python
#Numerical analysis
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
plt.figure(figsize=(18,12))
for i, col in enumerate(numerical_cols):
    plt.subplot(3,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

```python
#HEATMAPS
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot= True,cmap='crest',fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
## INSIGHTS OF EDA :
THE STRONGEST POSITIVE CORRELATION OCCUR AMONG "BANK DEPOSITS" WITH "CHECKING ACCOUNTS","SAVINGS ACCOUNTS" AND "FOREIGN CUURENCY ACCOUNTS" INDICATIONG THAT CUSTOMERS WHO MAINTAIN HIGH BALANCES IN ONE ACCOUNT TYPE OFTEN HOLD SUBSTANTIAL AMOUNT/FUNDS  ACROSS OTHER ACCOUNTS AS WELL
```
